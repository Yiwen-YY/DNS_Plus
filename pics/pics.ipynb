```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
service = "google.com"
name_list = []

for i in range(len(service)):
    name_list.append(service[0:i + 1])

name_list
```

```python
result_list = [73,
38,
33,
30,
30,
30,
24,
12,
8,
3
]

time_list = [
90.2694061,
72.4536144,
71.0106697,
71.6816791,
73.0388085,
70.95437270000001,
72.2750861,
74.00234209999999,
76.5489838,
78.792158
]
```

```python
sns.set()
plt.figure(figsize=(12, 3))
# plt.rcParams['font.family'] = 'Times New Roman'

value_list = np.linspace(0, 10, 10)

df = pd.DataFrame({'Name': name_list, 'Result': result_list,'Time': time_list, 'Value': value_list})

sns.lineplot(x='Value', y='Result', data=df, marker='o')

plt.xticks(ticks=value_list, labels=name_list)

# plt.title('Trend of Matched Results with Respect to Searched Name')
plt.xlabel('Searched Name')
plt.ylabel('Matched Results')
```

```python
plt.figure(figsize=(12, 3))

sns.lineplot(x='Value', y='Time', data=df, marker='o')

plt.xticks(ticks=value_list, labels=name_list)

# plt.title('Trend of Search Time with Respect to Searched Name')
plt.xlabel('Searched Name')
plt.ylabel('Search Time/ms')
```
